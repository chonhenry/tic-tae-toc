```python
from IPython.display import clear_output
clear_output()

def display_board(board):
    print(board[7] + '|' + board[8] + '|' + board[9])
    print('-+-+-')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-+-+-')
    print(board[1] + '|' + board[2] + '|' + board[3])
    
def player_input():
    print("Let's play tic tae toe")
    try:
        player_1 = input('Player 1 please select(X or O): ')
    except ValueError:
        player_1 = ' '
    
    while not (player_1.upper() == 'X' or player_1.upper() == 'O'):
        player_1 = input("Please don't enter anything other than X or O): ")
                         
    if player_1.upper()=='X':
        #return ('X','O')
        return 'X'
    else:
        #return ('O','X')
        return 'O'
    
def enter_position(board, turn):
    try:
        p = int(input(f'Player {turn} please enter a position(1-9): '))
    except ValueError:
        p = ' '
    
    while p not in range(1,10):
        try:
            p = int(input(f'Player {turn} please enter a position(1-9): '))
        except ValueError:
            p = ' '
        
    while board[p] != ' ':
        p = int(input(f'Position {p} has already been filled; please enter another position: '))
        
    return p
    
def place_marker(board, marker, position):
    board[position] = marker
    return board

def check_win(board):
    is_win = False
    
    if board[7] == board[8] == board[9] and board[7] != ' ':
        return(True)
    elif board[4] == board[5] ==board[6] and board[4] != ' ':
        return(True)
    elif board[1] == board[2] ==board[3] and board[1] != ' ':
        return(True)
    elif board[7] == board[4] ==board[1] and board[7] != ' ':
        return(True)
    elif board[8] == board[5] ==board[2] and board[8] != ' ':
        return(True)
    elif board[9] == board[6] ==board[3] and board[9] != ' ':
        return(True)
    elif board[7] == board[5] ==board[3] and board[7] != ' ':
        return(True)
    elif board[9] == board[5] ==board[1] and board[9] != ' ':
        return(True)
```

```python
#print("Let's play tic tae toe")
count=1
turn = 1
win = False
board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
marker = player_input()
player_marker = [' ',' ']
if marker == 'X':
    player_marker[0]='X'
    player_marker[1]='O'
else:
    player_marker[0]='O'
    player_marker[1]='X'
position = None

display_board(board)

while count <= 9:
    position = enter_position(board, turn)
    clear_output()
    board = place_marker(board, marker, position)
    display_board(board)
    
    win = check_win(board)
    if win == True:
        break
    
    if marker == 'X':
        marker = 'O'
    else:
        marker = 'X'
        
    if turn == 1:
        turn = 2
    else:
        turn = 1
    
    count+=1
    

print('')
if count == 9:
    print('tie game')
elif marker == 'X':
    print(f"Player {player_marker.index('X')+1} won!")
else:
    print(f"Player {player_marker.index('O')+1} won!")
```
